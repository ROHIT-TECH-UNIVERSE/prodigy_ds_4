# Start time

```python
ua.Start_Time
```

```python
#converting to timestamp type and override the original column
ua.Start_Time = pd.to_datetime(ua.Start_Time)
```

```python
#extracting  hour from timestamp 
ua.Start_Time.dt.hour
```

```python
# so as we have fetchn the data to make a visual representation
# we can now plot a histogram
ax=sns.distplot(ua.Start_Time.dt.hour, bins = 24 ,color='red',kde=False, norm_hist=True)
plt.grid(color="black")
ax.set(xlabel='Distribution of percentage of accident per day')
```

```python
#pull the day of the week from the timestamp format
###for monday

monday_start_time = ua.Start_Time[ua.Start_Time.dt.dayofweek == 0]
#plot the distribution
ax=sns.distplot(monday_start_time.dt.hour, bins = 24,color='orange', kde=False, norm_hist=True)
plt.grid(color="black")
ax.set(xlabel='Distribution of percentage of accident on monday')
```

```python
##fore sunday

sundays_start_time = ua.Start_Time[ua.Start_Time.dt.dayofweek == 6]
ax=sns.distplot(sundays_start_time.dt.hour, bins = 24,color='purple', kde=False, norm_hist=True)
plt.grid(color="black")
ax.set(xlabel='Distribution of percentage of accident on sunday')
```

```python
#### and at last we would se the time for overall month


sns.distplot(ua.Start_Time.dt.month, bins = 12, kde=False, norm_hist=True)
plt.grid(color="black")
```

```python
# as we have seen for in a day for monday and sunday  and month
# we would have a visualisation for a specific years
```

```python
###### 2016
ua_2016 = ua[ua.Start_Time.dt.year == 2016]
sns.distplot(ua_2016.Start_Time.dt.month, bins = 12,color='grey', kde=False, norm_hist=True)
plt.grid(color="red")
```

```python
#######2019
ua_2019 = ua[ua.Start_Time.dt.year == 2019]
sns.distplot(ua_2019.Start_Time.dt.month, bins = 12, kde=False,color='pink', norm_hist=True)
plt.grid(color="black")
```
